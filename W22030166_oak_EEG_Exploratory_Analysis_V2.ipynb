```python
from sklearn.preprocessing import StandardScaler

def standardize_data(df):
    # Initialize StandardScaler
    scaler = StandardScaler()
    standardized_data = scaler.fit_transform(df)

    # Create DataFrame from standardized data
    standardized_df = pd.DataFrame(standardized_data, columns=df.columns)

    return standardized_df
```

```python
shuffled_eeg_df.drop(['Zero Crossing Rate','Mean Absolute Value','Entropy'], axis=1, inplace=True)
```

```python
normalize_egg = standardize_data(shuffled_eeg_df.drop(['Epileptic_Seizure'], axis=1))
normalize_egg.head(10)
```

```python
processed_egg = pd.concat([normalize_egg, shuffled_eeg_df['Epileptic_Seizure']], axis=1)
processed_egg.head(10)
```

```python
import umap

sample=processed_egg.sample(n=20000, random_state=42)
X1 = sample.drop('Epileptic_Seizure', axis=1)
y1 = sample['Epileptic_Seizure']

reducer = umap.UMAP()
embedding = reducer.fit_transform(X1)
```

```python
# Plot the reduced data with class labels for coloring
plt.figure(figsize=(10, 8))
plt.scatter(embedding[y1 == 0, 0], embedding[y1 == 0, 1], label='Class 0', c='blue')
plt.scatter(embedding[y1 == 1, 0], embedding[y1 == 1, 1], label='Class 1', c='red')

plt.title('UMAP Visualization with Class Labels')
plt.xlabel('UMAP Dimension 1')
plt.ylabel('UMAP Dimension 2')
plt.legend()
plt.show()
```

```python
from sklearn.model_selection import train_test_split
seed=1234

X=processed_egg.loc[:,processed_egg.columns != 'Epileptic_Seizure']
y=processed_egg['Epileptic_Seizure']
def split_train_test(X, y, test_size):
    X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=test_size, random_state=42,shuffle=True, stratify=y)
        
    return X_train, X_test, y_train, y_test
```

```python
X_train, X_test, y_train, y_test=split_train_test(X, y, 0.2)
y_train.value_counts()
```
